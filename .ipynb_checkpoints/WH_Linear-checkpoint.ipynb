Project 3 Idea: What Factors Are Most Important To Happiness Based on 2017 World Happiness Report Data ðŸ¤ž

```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import os
WHR2017_df = pd.read_csv(os.path.join("data/csv/happy_2017.csv"))
WHR2017_df.head(5)
```

```python
# plt.savefig("TextFigurePlot.png")
```

```python
# WHR: DOES FAMILY MATTER?
# Yes. There is a clear visual linear pattern to the data. It says the happier you are with family, the happier you are.
# Which is not a shock. What we want to build toward is Which Factor is THE MOST impactful in our data.
# Seeing that a linear relationship appears to exist in the data should encourage us to fit a model to it and make
# predictions upon it. 
Family_X = WHR2017_df.HappinessScore.values.reshape(-1,1)
Family_y = WHR2017_df.Family.values.reshape(-1,1)
familyplot = plt.scatter(Family_X,Family_y)
```

```python
# WHR: DOES ECONOMY MATTER?

Econ_X = WHR2017_df.HappinessScore.values.reshape(-1,1)
Econ_y = WHR2017_df.EconomyGDPperCapita.values.reshape(-1,1)
econplot = plt.scatter(Econ_X,Econ_y)

```

```python
# WHR: DOES HealthLifeExpectancy MATTER?
Health_X = WHR2017_df.HappinessScore.values.reshape(-1,1)
Health_y = WHR2017_df.HealthLifeExpectancy.values.reshape(-1,1)
HealthLifeExpectancyplot=plt.scatter(Health_X,Health_y)
```

```python
# WHR: DOES FREEDOM MATTER?
Freedom_X = WHR2017_df.HappinessScore.values.reshape(-1,1)
Freedom_y = WHR2017_df.Freedom.values.reshape(-1,1)
Freedomplot = plt.scatter(Freedom_X,Freedom_y)
```

```python
# WHR: DOES GENEROSITY MATTER?
Generosity_X = WHR2017_df.HappinessScore.values.reshape(-1,1)
Generosity_y = WHR2017_df.Generosity.values.reshape(-1,1)
Generosityplot = plt.scatter(Generosity_X,Generosity_y)
```

```python
# WHR: DOES TrustGovernmentCorruption MATTER?
TrustGovt_X = WHR2017_df.HappinessScore.values.reshape(-1,1)
TrustGovt_y = WHR2017_df.TrustGovernmentCorruption.values.reshape(-1,1)
TrustGovtplot = plt.scatter(TrustGovt_X,TrustGovt_y)
```

```python
# WHR: DOES DystopiaResidual MATTER?
Dystopia_X = WHR2017_df.HappinessScore.values.reshape(-1,1)
Dystopia_y = WHR2017_df.DystopiaResidual.values.reshape(-1,1)
DystopiaResidualplot = plt.scatter(Dystopia_X,Dystopia_y)
```

Create Linear Regression Models, Fit the Models, Assign variables to hold coefficients and Y intercepts

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

FamilyX_train, FamilyX_test, Familyy_train, Familyy_test = train_test_split(Family_X, Family_y, random_state=1)
EconX_train, EconX_test, Econy_train, Econy_test = train_test_split(Econ_X, Econ_y, random_state=1)
HealthX_train, HealthX_test, Healthy_train, Healthy_test = train_test_split(Health_X, Health_y, random_state=1)
FreedomX_train, FreedomX_test, Freedomy_train, Freedomy_test = train_test_split(Freedom_X, Freedom_y, random_state=1)
GenerosityX_train, GenerosityX_test, Generosityy_train, Generosityy_test = train_test_split(Generosity_X, Generosity_y, random_state=1)
TrustGovtX_train, TrustGovtX_test, TrustGovty_train, TrustGovty_test = train_test_split(TrustGovt_X, TrustGovt_y, random_state=1)
DystopiaX_train, DystopiaX_test, Dystopiay_train, Dystopiay_test = train_test_split(Dystopia_X, Dystopia_y, random_state=1)
```

```python
Familymodel = LinearRegression()
Familymodel.fit(Familyy_train,FamilyX_train)
print(Familymodel.score(Familyy_test, FamilyX_test))

y_min = Family_y.min()
y_max = Family_y.max()

PredictedFamily_x_min = Familymodel.predict([[y_min]])
PredictedFamily_x_max = Familymodel.predict([[y_max]])

plt.scatter(Family_X,Family_y)
plt.plot([PredictedFamily_x_min[0],PredictedFamily_x_max[0]],[y_min,y_max],c="red")

# plt.savefig("TextFigurePlot.png")
```

```python
Econmodel = LinearRegression()
Econmodel.fit(Econy_train,EconX_train)
print(Econmodel.score(Econy_test, EconX_test))

y_min = Econ_y.min()
y_max = Econ_y.max()

PredictedEcon_x_min = Econmodel.predict([[y_min]])
PredictedEcon_x_max = Econmodel.predict([[y_max]])

plt.scatter(Econ_X,Econ_y)
plt.plot([PredictedEcon_x_min[0],PredictedEcon_x_max[0]],[y_min,y_max],c="red")
```

```python
Healthmodel = LinearRegression()
Healthmodel.fit(Healthy_train,HealthX_train)
print(Healthmodel.score(Healthy_test, HealthX_test))

y_min = Health_y.min()
y_max = Health_y.max()

PredictedHealth_x_min = Healthmodel.predict([[y_min]])
PredictedHealth_x_max = Healthmodel.predict([[y_max]])

plt.scatter(Health_X, Health_y)
plt.plot([PredictedHealth_x_min[0],PredictedHealth_x_max[0]],[y_min,y_max],c="red")
```

```python
Generositymodel = LinearRegression()
Generositymodel.fit(Generosityy_train,GenerosityX_train)
print(Generositymodel.score(Generosityy_test, GenerosityX_test))

y_min = Generosity_y.min()
y_max = Generosity_y.max()

PredictedGenerosity_x_min = Generositymodel.predict([[y_min]])
PredictedGenerosity_x_max = Generositymodel.predict([[y_max]])

plt.scatter(Generosity_X, Generosity_y)
plt.plot([PredictedGenerosity_x_min[0],PredictedGenerosity_x_max[0]],[y_min,y_max],c="red")
```

```python
Freedommodel = LinearRegression()
Freedommodel.fit(Freedomy_train,FreedomX_train)
print(Freedommodel.score(Freedomy_test, FreedomX_test))

y_min = Freedom_y.min()
y_max = Freedom_y.max()

PredictedFreedom_x_min = Freedommodel.predict([[y_min]])
PredictedFreedom_x_max = Freedommodel.predict([[y_max]])

plt.scatter(Freedom_X, Freedom_y)
plt.plot([PredictedFreedom_x_min[0],PredictedFreedom_x_max[0]],[y_min,y_max],c="red")
```

```python
Dystopiamodel = LinearRegression()
Dystopiamodel.fit(Dystopiay_train,DystopiaX_train)
print(Dystopiamodel.score(Dystopiay_test, DystopiaX_test))

y_min = Dystopia_y.min()
y_max = Dystopia_y.max()

PredictedDystopia_x_min = Dystopiamodel.predict([[y_min]])
PredictedDystopia_x_max = Dystopiamodel.predict([[y_max]])

plt.scatter(Dystopia_X, Dystopia_y)
plt.plot([PredictedDystopia_x_min[0],PredictedDystopia_x_max[0]],[y_min,y_max],c="red")
```
